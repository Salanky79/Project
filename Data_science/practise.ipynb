```python
# https://www.football-data.co.uk/mmz4281/2526/E1.csv
dict_historical_data = {}
dict_contries = {'premier_leage': 'E0', 'English league 1': 'E2', 'English league 2': 'E3'}
root = "https://www.football-data.co.uk/mmz4281/"
for league in dict_contries:
    frames = []
    for season in range(20,26):
        df = pd.read_csv( root + str(season) + str(season + 1) + "/" + dict_contries[league] + '.csv')
        df.insert(1, 'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```

```python
from IPython.display import display

for league, df in dict_historical_data.items():
    print(f"=== {league} ===")
    display(df)  # hiện bảng đẹp với đường kẻ
```
