```python
# Link examples:

# https://www.football-data.co.uk/mmz4281/2526/E0.csv
# https://www.football-data.co.uk/mmz4281/2526/E1.csv

# root + season + league



#create a root variable
root = "https://www.football-data.co.uk/mmz4281/"

#creating list of leagues
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []
#looping through leagues, read multiple csv and append it into a list (or else it will lost the df)
for league in leagues:
    df = pd.read_csv( root + '2526/' + league + '.csv', encoding='unicode_escape')
    frames.append(df)
```

```python
#-------------------------Multiple seasons----------------------



root = "https://www.football-data.co.uk/mmz4281/"

#creating list of leagues
leagues = ['E0', 'E2', 'E3']
frames = []
#looping through leagues, read multiple csv and append it into a list (or else it will lost the df)
for league in leagues:
    for season in range (20,26):
        df = pd.read_csv( root + str(season) + str(season+1) + '/' + league + '.csv', encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df) 
```

```python
len(frames)
```

```python
# --------------- Organize all the Data in a Dictionary ------------------
dict_historical_data = {}
dict_contries = {"English Premier League": 'E0', "English League 1": 'E2', "English League 2": 'E3'}

#creating list of leagues
#leagues = ['E0', 'E2', 'E3']

#looping through key elements
for league in dict_contries:
    frames = [] # ch·ª©a t·∫•t c·∫£ m√πa c·ªßa CH·ªà DUY NH·∫§T 1 GI·∫¢I (vd: t·∫•t c·∫£ c√°c m√πa (20-25) c·ªßa Engl√≠h league 1)
    for season in range (20,26):
        df = pd.read_csv( root + str(season) + str(season+1) + '/' + dict_contries[league] + '.csv', encoding='unicode_escape')
        df.insert(1, 'season', season)
        frames.append(df) 
    # üëâ N·ªëi (concat) t·∫•t c·∫£ DataFrame c·ªßa 1 gi·∫£i l·∫°i v·ªõi nhau
    df_concat = pd.concat(frames)
    # üëâ L∆∞u k·∫øt qu·∫£ v√†o dictionary
    dict_historical_data[league] = df_concat
```
