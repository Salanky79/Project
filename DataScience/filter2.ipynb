```python
import pandas as pd
import numpy as np

df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
```

```python
# set the cheapest laptop of each company first and the most expensive last
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
df_laptops
```

```python
# chi in ra cai loai re nhat cua moi cong ty
df_laptops.drop_duplicates(['Company'], keep='first', inplace=True, ignore_index=True)
df_laptops
```

```python
# chi in ra cai loai dat nhat cua moi cong ty
duplicated_first = df_laptops.duplicated('Company', keep='last')
df_laptops[~duplicated_first][['Company', 'Price_euros']] # ~ == not
```
