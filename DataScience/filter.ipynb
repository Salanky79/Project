```python
import numpy as np
price_tier = np.where(df_laptops['Price_euros'] > 2000, 'Expensive', 'Cheap')

df_laptops.drop(columns='Price', inplace=True)
df_laptops['Price_tier'] = price_tier
df_laptops.head() 
```

```python
df_laptops.value_counts('Price_tier')
```

```python
size_tier = np.where(df_laptops['Inches'] > 15, 'Big', 'Small')

idx = df_laptops.columns.get_loc('Inches')
df_laptops.drop(columns='Size_tier', inplace=True)
df_laptops.insert(idx + 1, 'Size_tier', size_tier)
df_laptops.value_counts('Size_tier')
```

```python
# -------------------------- Multiple coditions ---------------------------------
df_laptops[(df_laptops['Company'] == "Apple") & (df_laptops['Price_euros'] > 2000)] 
```

```python
conditions = [
    df_laptops['Price_euros'] > 3000,
    (df_laptops['Price_euros'] > 2000) & (df_laptops['Price_euros'] <= 3000),
    (df_laptops['Price_euros'] > 800) & (df_laptops['Price_euros'] <= 2000),
    df_laptops['Price_euros'] <= 800

]

values = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
#set it to new column
df_laptops['Price_tier'] = np.select(conditions, values, default='Unknown')
df_laptops
```
